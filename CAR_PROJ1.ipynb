```python

df['Engine Cylinders'].value_counts().plot(kind = 'bar')

plt.xticks(rotation = 0)
plt.xlabel('AUTOMATIC')
plt.ylabel('Large')
plt.title('car analysis')
plt.grid(True)
plt.show()
```

```python
plt.plot(df['Engine Cylinders'] , label = 'Engine Cylinders' , color = 'black')
plt.plot(df['Engine HP'] , label = 'Engine HP', color = 'green')
plt.legend()

plt.show()
```

```python
# making subplot
plt.figure(figsize = (20,3))
plt.subplot(1,2,1)
plt.plot(df['Engine Cylinders'])
plt.legend()
plt.subplot(1,2,2)
plt.plot(df['Engine HP'], color = 'red')
plt.legend()
plt.show()
```
