```python
# lets get into research attribute
RES_Count = data.groupby(['Research']).count()
RES_Count = RES_Count['GRE Score']
RES_Count = pd.DataFrame(RES_Count)
RES_Count.rename({'GRE Score': 'Count'}, axis=1, inplace=True)
RES_Count.rename({0: 'No Research', 1:'Research'}, axis=0, inplace=True)
plt.pie(x=RES_Count['Count'], labels=RES_Count.index, autopct='%1.1f%%')
plt.title('Research', pad=10, size=15)
plt.show()
```

```python
# Barplot for the distribution of the University Rating
University_Rating = data.groupby(['University Rating']).count()
University_Rating = University_Rating['GRE Score']
University_Rating = pd.DataFrame(University_Rating)
University_Rating.rename({'GRE Score': 'Count'}, inplace=True, axis=1)
sns.barplot(University_Rating.index, University_Rating['Count']).set_title('University Rating', size='15')
plt.show()
```

```python
data = pd.read_csv('Admission_predict.CSV',index_col=0)
data = data.rename(columns={'GRE Score' : 'GRE_Score',
                            'TOEFL Score' : 'TOEFL_Score',
                            'University Rating' : 'University_Rating',
                            'Chance of Admit ' : 'Chance_of_Admit',
                            'LOR ' : 'LOR'})

data.loc[data.Chance_of_Admit >= 0.90, 'Chance_of_Admit_Class'] = "High"
data.loc[(data.Chance_of_Admit >= 0.75) & (data.Chance_of_Admit < 0.90), 'Chance_of_Admit_Class'] = "Average"
data.loc[data.Chance_of_Admit <= 0.75, 'Chance_of_Admit_Class'] = "Low"

del data['Chance_of_Admit']

print (data['Chance_of_Admit_Class'].value_counts())
```

```python
plt.figure(1, figsize=(10,6))
plt.subplot(1,4, 1)
plt.boxplot(data['GRE_Score'])
plt.title('GRE_Score')

plt.subplot(1,4,2)
plt.boxplot(data['TOEFL_Score'])
plt.title('TOEFL_Score')

plt.subplot(1,4,3)
plt.boxplot(data['University_Rating'])
plt.title('University_Rating')

plt.subplot(1,4,4)
plt.boxplot(data['CGPA'])
plt.title('CGPA')

plt.show()
```
