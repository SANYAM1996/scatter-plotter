```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Admission_predict.CSV')
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.corr()
```

```python
# correlation heat map
sns.heatmap(data.corr(), annot=True).set_title('Correlation Factors Heat Map', color='blue', size='15')
```

```python
# lets see Gre score dataset
GRE = pd.DataFrame(data['GRE Score'])
GRE.describe()
```

```python
# previously GRE mean score is 306 now its 316 which is bit higher
# lets see probability of GRE score
sns.distplot(GRE).set_title('Probability Distribution for GRE Test Scores', size='15')
plt.show()
```

```python
# lets move to TOFEL score
TOEFL = pd.DataFrame(data['TOEFL Score'], columns=['TOEFL Score'])
TOEFL.describe()
```

```python
sns.distplot(TOEFL).set_title('Probability Distribution for TOEFL Scores', size='15')
plt.show()
```

```python
# Describing the CGPA
CGPA = pd.DataFrame(data['CGPA'], columns=['CGPA'])
CGPA.describe()
```

```python
sns.distplot(CGPA).set_title('Probability Distribution Plot for CGPA', size='15')
plt.show()
```

```python
# lets get into research attribute
RES_Count = data.groupby(['Research']).count()
RES_Count = RES_Count['GRE Score']
RES_Count = pd.DataFrame(RES_Count)
RES_Count.rename({'GRE Score': 'Count'}, axis=1, inplace=True)
RES_Count.rename({0: 'No Research', 1:'Research'}, axis=0, inplace=True)
plt.pie(x=RES_Count['Count'], labels=RES_Count.index, autopct='%1.1f%%')
plt.title('Research', pad=10, size=15)
plt.show()
```

```python
# Barplot for the distribution of the University Rating
University_Rating = data.groupby(['University Rating']).count()
University_Rating = University_Rating['GRE Score']
University_Rating = pd.DataFrame(University_Rating)
University_Rating.rename({'GRE Score': 'Count'}, inplace=True, axis=1)
sns.barplot(University_Rating.index, University_Rating['Count']).set_title('University Rating', size='15')
plt.show()
```

```python
# Describing sop
SOP = pd.DataFrame(data.groupby(['SOP']).count()['GRE Score'])
SOP.rename({'GRE Score':'Count'}, axis=1, inplace=True)
SOP
```

```python
sns.barplot(SOP.index, SOP['Count']).set_title('Statement of Purpose', size='15')
plt.show()
```
