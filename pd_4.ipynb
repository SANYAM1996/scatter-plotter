```python
%pylab inline
import pandas
import seaborn
```

```python
data = pandas.read_csv("uber-raw-data-apr14.csv")
data
```

```python
data.tail(50)
```

```python
dt = '4/30/2014 21:34:00'
dt
```

```python
d, t = dt.split(' ')
print(d)
print(t)
```

```python
dt = pandas.to_datetime(dt)
```

```python
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data.tail()
```

```python
def get_dom(dt):
    return dt.day

data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.head(100)
```

```python
def get_weekday(dt):
    return dt.weekday
data['weekday'] = data['Date/Time'].map(get_weekday)
def get_hour(dt):
    return dt.hour
data['hour'] = data['Date/Time'].map(get_hour)
data.head(100)
```

```python
data.describe()
```

### ANAYSIS

```python
hist (data.dom, bins = 30, rwidth=.5, color = 'black',range = (0.5, 30.5))
xlabel('date of the month')
ylabel('frequency')
title('uber-april-2014')
```

```python
for k, rows in data.groupby('dom'):
    print ((k, rows))
    break
```

```python
def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

## analysing the data according to date

```python
plot (by_date,color = 'indigo' )
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

## analysing the sorted date 

```python
hist(by_date_sorted,rwidth =.7,color = 'red')

xlabel('date of the month')
ylabel('frequency')
title('uber-april-2014')
```
