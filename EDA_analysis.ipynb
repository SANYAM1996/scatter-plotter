```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('crime.csv',encoding="cp1252")
data
```

```python
# find out the most frequent crimes
data.columns=[ x.lower() for x in data.columns]
data.columns
crime_most_common=data['offense_code_group'].value_counts()
crime_most_common.head()



crime_most_common=crime_most_common.to_frame()
crime_most_common.columns

plt.figure(figsize=(12, 6))

crime_most_common.head()


sns.barplot(x=crime_most_common.index[:5],y="offense_code_group",data=crime_most_common.head(),orient="v")
plt.title("Frequency of top five crime committed in Boston",fontsize=30)
```

```python
#crime frequency year, week , month
data.head()

crime_by_days=data[['offense_code_group','year','month','day_of_week','hour','occurred_on_date',]]
crime_by_days
#data.occurred_on_date.dtypes
plt.figure(figsize=(10, 6))

sns.countplot(x="day_of_week",data=crime_by_days)
plt.title("Frequency of  crime committed in Boston in a week",fontsize=30)
```

```python
#month
plt.figure(figsize=(8, 6))

sns.countplot(x="month",data=crime_by_days)
plt.title("Frequency of  crime committed in Boston across the month",fontsize=30)
```

```python
#year
plt.figure(figsize=(8, 6))

sns.countplot(x="year",data=crime_by_days)
plt.title("Frequency of  crime committed in Boston across the years",fontsize=20)
```

```python
#specify thetype of crime from the above list 
def crime_specified(x):
    crime_specified=data[data["offense_code_group"]==x]
    plt.figure(figsize=(16,6))
    sns.countplot(x="district",data=crime_specified)
    plt.title("Frequency of "+str(x)+" "+"response across the districts:", fontsize=15)
    return crime_specified
```

```python
#district wise stats of the distribution
def street_wise_district(x):   
    crime_district=data[data["district"]==x]
    
    
    plot_fig(crime_district,x,"offense_code_group")
    plot_fig(crime_district,x,"street")
    plot_time(crime_district,x,"day_of_week")
    plot_time(crime_district,x,"hour")
    plot_time(crime_district,x,"month")
    return crime_district
```

```python
def plot_fig(crime,string,y):
    x=crime[y].value_counts()
    x=x.to_frame()
    
    plt.figure(figsize=(16,6))
    sns.barplot(x=x.index[:5],y=y,data=x.head(),orient="v")
    plt.title("Frequency of top 5 crime committed in Boston in "+str(y)+" of "+str(string),fontsize=15)
```

```python
def plot_time(crime_district,x,time):
    plt.figure(figsize=(16, 6))
    sns.countplot(x=time,data=crime_district)
    plt.title("Frequency of  crime committed in Boston in "+str(x)+" for "+str(time),fontsize=15)
```
