```python
data.plot(kind = 'Scatter',x='Sepal length', y = 'Sepal width');
plt.grid(True)
plt.show()
sns.set_style("dark")
```

```python
sns.set_style("darkgrid")
sns.FacetGrid(data, hue= "Species") \
   .map(plt.scatter,"Sepal length", "Sepal width") \
   . add_legend();
plt.show()
```

```python
plt.close()
```

```python
sns.set_style("whitegrid");
sns.pairplot(data, hue="Species",size=3);
plt.show()
```

```python
# by studying above pair plots in petal length row 4th column its kinda easy to understand the species of flower because its the only one plot in which the data is distincly separated
# i am creating a small model here of every species with "if else command"
# here i am taking x=petal length and y = petal width
```

```python
x = 1
y = 1
if (x<2) & (y<2):
    print("Species is setosa")
```

```python
x = 5
y = 2
if (x<=6) & (y<=3):
    print("Species is versicolor")
```

```python
x = 7
y = 3
if (x<=8) & (y<=4):
    print("Species is verginica")
```

```python
sns.FacetGrid(data, hue= "Species",size = 5) \
   .map(plt.hist, "Petal length") \
   . add_legend();
plt.grid(True)
plt.show()
```

```python
sns.FacetGrid(data, hue= "Species",size = 5) \
   .map(plt.hist, "Petal width") \
   . add_legend();
plt.grid(True)
plt.show()
```

```python
sns.FacetGrid(data, hue= "Species",size = 5) \
   .map(plt.hist, "Sepal length") \
   . add_legend();
plt.grid(True)
plt.show()
```

```python
sns.FacetGrid(data, hue= "Species",size = 5) \
   .map(plt.hist, "Sepal width") \
   . add_legend();
plt.grid(True)
plt.show()
```

```python
# after univariate analysing the above graph we conclude that petal length>>petal width>>Sepal length>>Sepal width
# petal length has more significant value rather than others to distinguis the among species because its well separated and have less 
# overlapping that other dimensions.
```
