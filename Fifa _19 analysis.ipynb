```python
# lets choose columns in which we are intrested
chosen_columns = [
    'Name',
    'Age',
    'Nationality',
    'Overall',
    'Potential',
    'Special',
    'Acceleration',
    'Aggression',
    'Agility',
    'Balance',
    'BallControl',
    'Body Type',
    'Composure',
    'Crossing',
    'Curve',
    'Club',
    'Dribbling',
    'FKAccuracy',
    'Finishing',
    'GKDiving',
    'GKHandling',
    'GKKicking',
    'GKPositioning',
    'GKReflexes',
    'HeadingAccuracy',
    'Interceptions',
    'International Reputation',
    'Jersey Number',
    'Jumping',
    'Joined',
    'LongPassing',
    'LongShots',
    'Marking',
    'Penalties',
    'Position',
    'Positioning',
    'Preferred Foot',
    'Reactions',
    'ShortPassing',
    'ShotPower',
    'Skill Moves',
    'SlidingTackle',
    'SprintSpeed',
    'Stamina',
    'StandingTackle',
    'Strength',
    'Value',
    'Vision',
    'Volleys',
    'Wage',
    'Weak Foot',
    'Work Rate'
]
```

```python
# i created new dataframe with above columns
df=pd.DataFrame(data,columns=chosen_columns)
```

```python
df.sample(10)
```

```python
import seaborn as sns
```

```python
# lets create correlation heat map
plt.rcParams['figure.figsize']=(30,20)
hm=sns.heatmap(df[['Age', 'Overall', 'Potential', 'Value', 'Wage',
                'Acceleration', 'Aggression', 'Agility', 'Balance', 'BallControl', 
                'Body Type','Composure', 'Crossing','Dribbling', 'FKAccuracy', 'Finishing', 
                'HeadingAccuracy', 'Interceptions','International Reputation',
                'Joined', 'Jumping', 'LongPassing', 'LongShots',
                'Marking', 'Penalties', 'Position', 'Positioning',
                'ShortPassing', 'ShotPower', 'Skill Moves', 'SlidingTackle',
                'SprintSpeed', 'Stamina', 'StandingTackle', 'Strength', 'Vision',
                'Volleys']].corr(), annot = True, linewidths=.5, cmap='gray')
hm.set_title(label='Heatmap of dataset', fontsize=20)
hm;
```

```python
# Histogram: number of players's age
sns.set(style ="dark", palette="colorblind", color_codes=True)
x = df.Age
plt.figure(figsize=(12,8))
ax = sns.distplot(x, bins = 58, kde = False, color='r')
ax.set_xlabel(xlabel="Player\'s age", fontsize=16)
ax.set_ylabel(ylabel='Number of players', fontsize=16)
ax.set_title(label='Histogram of players age', fontsize=20)
plt.grid(True)
plt.show()
```

```python
# making scatter plotter between accleration and other chosen field
def A (df):
    feats = ('Agility','Balance','Dribbling','SprintSpeed')
    
    for index, feat in enumerate(feats):
        plt.subplot(len(feats)/4+1,4,index+1)
        ax = sns.regplot(x= 'Acceleration',y = feat,data = df)
plt.figure(figsize=(20,20))
plt.subplots_adjust(hspace = 0.4)
A(df)
```

```python
# lets find out the elder players in dataset
elder = df.sort_values('Age',ascending = False)[['Name','Nationality','Age']].head(20)
elder.set_index('Name',inplace=True)
print(elder)
```

```python
# lets find out the youngest players in dataset
youngest = df.sort_values('Age',ascending = True)[['Name','Nationality','Age']].head(20)
youngest.set_index('Name',inplace=True)
print(youngest)
```

```python
# Compare six clubs in relation to age
some_clubs = ('Juventus', 'Real Madrid', 'Manchester City', 'FC Barcelona', 'Paris FC', 'Manchester United')
df_club = df.loc[df['Club'].isin(some_clubs) & df['Age']]

fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
ax = sns.violinplot(x="Club", y="Age", data=df_club);
ax.set_title(label='Distribution of age in some clubs', fontsize=20);
```

```python
# lets see the oldest team
df.groupby(['Club'])['Age'].sum().sort_values(ascending = False).head(20)
```

```python
# lets see the youngest team
df.groupby(['Club'])['Age'].sum().sort_values(ascending = True).head(20)
```

```python
# The clubs and their players overalls
some_clubs = ('Juventus', 'Real Madrid', 'Manchester City', 'FC Barcelona', 'Paris FC', 'Manchester United')
df_club = df.loc[df['Club'].isin(some_clubs) & df['Age'] & df['Overall'] ]
plt.grid(True)

ax = sns.barplot(x=df_club['Club'], y=df_club['Overall'], palette="deep");
ax.set_title(label='Distribution overall in several clubs', fontsize=25);
```

```python
# All of position
ax = sns.countplot(x = 'Position', data = df, palette = 'deep');
ax.set_title(label='Count of players on the position', fontsize=20);
plt.grid(True)
plt.show();
```
