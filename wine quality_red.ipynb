```python
data['quality'].value_counts()
```

```python
data['quality'].value_counts().plot(kind = 'bar')
plt.grid(True)
plt.show()
```

```python
# values rotated on x bar
data['quality'].value_counts().plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.grid(True)
plt.show()
```

```python
# sorted values
sorted_values = data['quality'].value_counts().sort_index()
sorted_values.plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.grid(True)
plt.show()
```

```python
sorted_values.plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.xlabel('quality score')
plt.ylabel('count')
plt.title('wine quality')
plt.grid(True)
plt.show()
```
