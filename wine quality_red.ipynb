### seaborn:correlation heatmap

```python
corr = data.corr()
corr
```

```python
import numpy as np
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = False

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=False)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

## seaborn : pairplot

```python
sns.distplot(data['quality'])
plt.show()
```

```python
# Bin "quality" variable into three levels: poor, normal and excellent

bins = [0, 4, 6, 10]
labels = ["poor","normal","excellent"]
data['binned_quality'] = pd.cut(data['quality'], bins=bins, labels=labels)
data.head()
data.drop('quality',axis =1, inplace = True)
```

```python
sns.distplot(data['alcohol'], color = 'red')
plt.show()
```

```python
# Bin "alcohol" variable into three levels: low, median and high

bins = [0, 10, 12, 15]
labels = ["low alcohol","median alcohol","high alcohol"]
data['binned_alcohol'] = pd.cut(data['alcohol'], bins=bins, labels=labels)
data.drop('alcohol',axis =1, inplace = True)
```

```python
data.head()
```

```python
sns_plot = sns.pairplot(data, hue="binned_quality", palette="husl",
             diag_kind="kde")
sns_plot.savefig("pairplot.png")
```
