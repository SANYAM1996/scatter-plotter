```python
data['quality'].value_counts().plot(kind = 'bar')
plt.grid(True)
plt.show()
```

```python
# values rotated on x bar
data['quality'].value_counts().plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.grid(True)
plt.show()
```

```python
# sorted values
sorted_values = data['quality'].value_counts().sort_index()
sorted_values.plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.grid(True)
plt.show()
```

```python

data['quality'].value_counts().plot(kind = 'bar')

plt.xticks(rotation = 0)
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.title('wine quality')
plt.grid(True)
plt.show()
```

```python
# making subplot
plt.subplot(1,2,1)
plt.plot(data['fixed acidity'])
plt.subplot(1,2,2)
plt.plot(data['pH'], color = 'red')
plt.show()

```

```python
plt.figure(figsize = (20,3))
plt.subplot(1,2,1)
plt.plot(data['fixed acidity'])
plt.subplot(1,2,2)
plt.plot(data['pH'], color = 'red')
plt.show()

```

```python

sns.jointplot(x='fixed acidity',y = 'pH',data = data)
plt.show()
```

```python
# 3rd order regression line shows the corellaion between both two label axis.
sns.regplot(x='fixed acidity',y = 'pH',data = data,order = 3)
plt.show()
```
