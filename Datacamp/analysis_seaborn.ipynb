```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x = 'total_bill',y='tip',kind = 'scatter',data = tips)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill',y='tip',hue = 'smoker',data = tips)
plt.show()
```

```python
# setting hue color
hue_colors = {'Yes':'black',
              'No':'blue'}
sns.scatterplot(x = 'total_bill',y='tip',hue = 'smoker',data = tips,palette=hue_colors)
plt.show()
```

```python
# subplots in column
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter',col = 'smoker')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter',row = 'smoker')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter',row = 'smoker',col = 'time',hue = 'day')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter',col = 'day',hue = 'time',col_wrap = 2)
```

```python
# subgroups with point size
sns.relplot(x='total_bill',y ='tip',data = tips,kind = 'scatter',size = 'size',hue = 'size')
```

```python
sns.relplot(x='total_bill',y ='tip',data = tips,kind = 'scatter',style = 'smoker',hue = 'size',)
```
