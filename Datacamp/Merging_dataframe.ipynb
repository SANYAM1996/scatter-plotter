```python
# Arithmetic with series and dataframes
df = pd.read_csv('pittsburgh2013.csv',index_col='Date', parse_dates=True)
df
```

```python
df.loc['2013-7-1':'2013-7-7', 'PrecipitationIn']
```

```python
# scalar multiplication
df.loc['2013-07-01':'2013-07-07', 'PrecipitationIn'] * 7.54
```

```python
# absolute temperature range
week1_range = df.loc['2013-07-01':'2013-07-07',['Min TemperatureF', 'Max TemperatureF']]
```

```python
week1_range
```

```python
# average temperature
week1_mean = df.loc['2013-07-01':'2013-07-07','Mean TemperatureF']
```

```python
week1_mean
```

```python
week1_range.divide(week1_mean, axis=0)
```

```python
# percentage change
week1_mean.pct_change() * 100
```

# using append and concate in dataset

```python
# Using .append()
northeast = pd.Series(['CT', 'ME', 'MA', 'NH', 'RI', 'VT','NJ', 'NY', 'PA'])
south = pd.Series(['DE', 'FL', 'GA', 'MD', 'NC', 'SC', 'VA','DC', 'WV', 'AL', 'KY', 'MS', 'TN', 'AR', 'LA', 'OK', 'TX'])
midwest = pd.Series(['IL', 'IN', 'MN', 'MO', 'NE', 'ND','SD', 'IA', 'KS', 'MI', 'OH', 'WI'])
west = pd.Series(['AZ', 'CO', 'ID', 'MT', 'NV', 'NM','UT', 'WY', 'AK', 'CA', 'HI', 'OR','WA'])
east = northeast.append(south)
east
```

```python
# The appended Index
east.index
```

```python
east.loc[3]
```

```python
# Using .reset_index()
new_east = northeast.append(south).reset_index(drop=True)
print(new_east.head(11))
```

```python
print(new_east.index)
```

```python
# Using concat()
east = pd.concat([northeast, south])
```

```python
east.head(11)
```

```python
print(east.index)
```

```python
# Using ignore_index
new_east = pd.concat([northeast, south],ignore_index=True)
```

```python
print(new_east.head(11))
```

```python
print(new_east.index)
```
