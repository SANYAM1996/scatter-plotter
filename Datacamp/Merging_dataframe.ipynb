```python
# using a loop to call file
filenames = ['sales-jan-2015.csv', 'sales-feb-2015.csv']
dataframes = []
for filename in filenames:
    dataframes.append(pd.read_csv(filename))
    
print(dataframes[0].head(7))
```

```python
# using comprehension
filenames = ['sales-jan-2015.csv', 'sales-feb-2015.csv']
dataframes = [pd.read_csv(filename) for filename in filenames]
dataframes
```

```python
# Arithmetic with series and dataframes
df = pd.read_csv('pittsburgh2013.csv',index_col='Date', parse_dates=True)
df
```

```python
df.loc['2013-7-1':'2013-7-7', 'PrecipitationIn']
```

```python
# scalar multiplication
df.loc['2013-07-01':'2013-07-07', 'PrecipitationIn'] * 7.54
```

```python
# absolute temperature range
week1_range = df.loc['2013-07-01':'2013-07-07',['Min TemperatureF', 'Max TemperatureF']]
```

```python
week1_range
```

```python
# average temperature
week1_mean = df.loc['2013-07-01':'2013-07-07','Mean TemperatureF']
```

```python
week1_mean
```

```python
week1_range.divide(week1_mean, axis=0)
```

```python
# percentage change
week1_mean.pct_change() * 100
```
