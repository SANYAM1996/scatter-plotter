```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('bigcity.csv')
df
```

```python
# creating a scatter plot
plt.scatter(x = 'u',y = 'x',data = df,marker = 's',c = 'darkred')
```

```python
# include edge color
plt.scatter(x = 'x',y = 'u',data = df,edgecolor = 'darkblue',c = 'yellow')
```

```python
# add grid lines and axis label

plt.scatter(x = 'x',y = 'u',data = df,edgecolor = 'darkblue',c = 'yellow')
plt.grid
plt.xlabel('x bar')
plt.ylabel('u bar')
plt.title('big city')

plt.show()
```
