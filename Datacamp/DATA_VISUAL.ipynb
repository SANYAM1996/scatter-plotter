```python
climate_change = pd.read_csv('env-data-del.csv',parse_dates=["day"],index_col="day")
```

```python
climate_change
```

```python
# Plotting two time-series together
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["wind speed"])
ax.plot(climate_change.index, climate_change["temperature"])
ax.set_xlabel('Time')
ax.set_ylabel('wind speed / temperature')
plt.show()
```

```python
# Using twin axes
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["wind speed"])
ax.set_xlabel('Time')
ax.set_ylabel('wind speed')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["temperature"])
ax2.set_ylabel('temperature')
plt.show()
```

```python
# Separating variables by color
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["wind speed"], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('wind speed', color='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["temperature"],
color='red')
ax2.set_ylabel('temperature', color='red')
plt.show()
```

```python
# Coloring the ticks
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["wind speed"],
color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('wind speed', color='blue')
ax.tick_params('y', colors='blue')
ax2 = ax.twinx()
ax2.plot(climate_change.index,
climate_change["temperature"],
color='red')
ax2.set_ylabel('temperature',
color='red')
ax2.tick_params('y', colors='red')
plt.show()
```
