```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sp500.csv')
data = df.head()
data
```

```python
plt.plot(data['Open'],data['Close'])
plt.plot(data['Low'],data['High'])
plt.show()
```

```python
# customization
```

```python
plt.plot(data['Open'],data['Close'],marker = 's',linestyle = '--')
```

```python
plt.plot(data['Low'],data['High'],marker = 'd',color = 'r',linestyle = 'none')
```

```python
plt.plot(data['Open'],data['Close'],marker = 's',linestyle = '--')
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Open vs close')
plt.show()
```

```python
# adding more data
plt.plot(data['Open'],data['Close'],marker = 's',linestyle = '--')
plt.plot(data['Low'],data['High'])
plt.plot(data['Volume'],data['Adj Close'],marker = 'o',linestyle = '--')
plt.show()
```
